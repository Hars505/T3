```python
class Student:
    def __init__(self, name, roll_no, enrollment_no, branch, email, contact_no):
        self.name = name
        self.roll_no = roll_no
        self.enrollment_no = enrollment_no
        self.branch = branch
        self.email = email
        self.contact_no = contact_no
        self.registered_events = []

    def display_details(self):
        print(f"\nStudent Details:")
        print(f"Name: {self.name}")
        print(f"Roll No: {self.roll_no}")
        print(f"Enrollment No: {self.enrollment_no}")
        print(f"Branch: {self.branch}")
        print(f"Email: {self.email}")
        print(f"Contact No: {self.contact_no}")

    def display_registered_events(self):
        if self.registered_events:
            print(f"\nRegistered Events for {self.name} ({self.enrollment_no}):")
            for event in self.registered_events:
                print(f"- {event}")
        else:
            print(f"\n{self.name} ({self.enrollment_no}) has not registered for any events.")


class EventSelection:
    def __init__(self, student):
        self.student = student
        self.group_events = ["Cricket", "Football", "Basketball", "Volleyball"]
        self.individual_events = ["Sprint Race", "Table Tennis"]
        self.max_group_events = 2

    def display_available_events(self):
        print("\nAvailable Events:")
        print("Group Events:")
        for event in self.group_events:
            print(f"- {event}")
        print("\nIndividual Events:")
        for event in self.individual_events:
            print(f"- {event}")

    def select_event(self, event_name):
        if event_name in self.group_events:
            if len(self.student.registered_events) < self.max_group_events:
                self.student.registered_events.append(event_name)
                print(f"{event_name} selected successfully!")
            else:
                print(f"Error: Maximum limit of {self.max_group_events} group events reached.")
        elif event_name in self.individual_events:
            self.student.registered_events.append(event_name)
            print(f"{event_name} selected successfully!")
        else:
            print(f"Error: {event_name} is not a valid event.")


# Menu
def display_menu():
    print("\nMenu:")
    print("1. Register a new student")
    print("2. Register for events")
    print("3. Search by Enrollment No")
    print("4. Exit")


def register_new_student():
    name = input("Enter student name: ")
    roll_no = input("Enter roll number: ")
    enrollment_no = input("Enter enrollment number: ")
    branch = input("Enter branch: ")
    email = input("Enter email: ")
    contact_no = input("Enter contact number: ")

    student = Student(name, roll_no, enrollment_no, branch, email, contact_no)
    student.display_details()
    return student


def register_for_events(students):
    enrollment_no = input("Enter enrollment number to register for events: ")
    student_found = False

    for student in students:
        if student.enrollment_no == enrollment_no:
            student_found = True
            event_selection = EventSelection(student)
            event_selection.display_available_events()

            while True:
                event_name = input("\nEnter an event to select (or 'done' to finish): ")
                if event_name.lower() == 'done':
                    break
                event_selection.select_event(event_name)

            student.display_registered_events()
            break

    if not student_found:
        print("Student not found.")


def search_by_enrollment(students):
    enrollment_no = input("Enter enrollment number to search: ")
    student_found = False

    for student in students:
        if student.enrollment_no == enrollment_no:
            student_found = True
            student.display_details()
            student.display_registered_events()
            break

    if not student_found:
        print("Student not found.")


# Main program loop
students = []

while True:
    display_menu()
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        student = register_new_student()
        students.append(student)
    elif choice == '2':
        register_for_events(students)
    elif choice == '3':
        search_by_enrollment(students)
    elif choice == '4':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```
