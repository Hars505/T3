```python
# WAP to create a class book which has 2 methods fiction and non-fiction both the methods are created to genrate an inventory of book also 
#create a method to display this books
class Book:
    def __init__(self):
        self.fict=list()
        self.nfict=list()
    def fictional(self,name,author):
        self.name=name
        self.author=author
        self.fict+=["Name : "+self.name, "Author : "+self.author]
    def non_fictional(self,name,author):
        self.name=name
        self.author=author
        self.nfict+=["Name : "+self.name, "Author : "+self.author]
    def display(self):
        print("\nfictional books\n")
        for i in self.fict: 
            print(i)
        print("\nNon fictional books\n")
        for i in self.nfict: 
            print(i)
```

```python
b1=Book()
b1.non_fictional("Wings of fire","Ravindranath Tagore")
b1.fictional("Hell's Paradise","Kijmi Nakamoto")
b1.fictional("My last Moments","Louis Stepherd")
```

```python
b1.display()
```

#### WAP to create a class student which stores name of student and marks of student in constructor create to separate,methods
#### one for entering marks of 3 different subjects and display the marksheet

```python
class Student:
    def __init__(self,name,rn):
        self.name=name
        self.rn=rn
        print("Enter marks")
        self.marks(int(input("Enter marks of Python : ")),int(input("Enter marks of F.S.D. : ")),int(input("Enter marks of D.E. : ")))
    def marks(self,pyh,fsd,de):
        self.pyh=pyh
        self.fsd=fsd
        self.de=de
    def marksht(self):
        print("-------------------------------------------------")
        print(f"Name Of Student         : {self.name}")
        print(f"Roll Number of Student  :{self.rn}")
        print("-------------------------------------------------")
        print(f"Marks of Python     : {self.pyh}")
        print(f"Marks of D.E.       : {self.de}")
        print(f"Marks of F.S.D.     : {self.fsd}")
        print("-------------------------------------------------")
        per=round(((self.pyh+self.de+self.fsd)/75)*100,2)
        print(f"Total Percentage Of Three Subjects : {per} ")
        if (per>=70):
            print("congratulations - You got Distinction")
        elif (per<=70 and per>=40):
            print("congratulations - You got Pass")
        else:
            print("You have Failed")
        print("-------------------------------------------------")
```

```python
stu1=Student(input("Enter Student name : "),input("Enter Student Roll Number : "))
stu2=Student(input("Enter Student name : "),input("Enter Student Roll Number : "))
```

```python
stu1.marksht()
```

```python
stu2.marksht()
```
