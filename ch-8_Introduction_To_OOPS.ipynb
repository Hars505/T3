```python
# WAP to create a class book which has 2 methods fiction and non-fiction both the methods are created to genrate an inventory of book also 
#create a method to display this books
class Book:
    def __init__(self):
        self.fict=list()
        self.nfict=list()
    def fictional(self,name,author):
        self.name=name
        self.author=author
        self.fict+=["Name : "+self.name, "Author : "+self.author]
    def non_fictional(self,name,author):
        self.name=name
        self.author=author
        self.nfict+=["Name : "+self.name, "Author : "+self.author]
    def display(self):
        print("\nfictional books\n")
        for i in self.fict: 
            print(i)
        print("\nNon fictional books\n")
        for i in self.nfict: 
            print(i)
```

#### WAP to create a class student which stores name of student and marks of student in constructor create to separate,methods
#### one for entering marks of 3 different subjects and display the marksheet

```python
class Student:
    def __init__(self,name,rn):
        self.name=name
        self.rn=rn
        print("Enter marks")
        self.marks(int(input("Enter marks of Python : ")),int(input("Enter marks of F.S.D. : ")),int(input("Enter marks of D.E. : ")))
    def marks(self,pyh,fsd,de):
        self.pyh=pyh
        self.fsd=fsd
        self.de=de
    def marksht(self):
        print("-------------------------------------------------")
        print(f"Name Of Student         : {self.name}")
        print(f"Roll Number of Student  : {self.rn}")
        print("-------------------------------------------------")
        print(f"Marks of Python     : {self.pyh}")
        print(f"Marks of D.E.       : {self.de}")
        print(f"Marks of F.S.D.     : {self.fsd}")
        print("-------------------------------------------------")
        per=round(((self.pyh+self.de+self.fsd)/75)*100,2)
        print(f"Total Percentage Of Three Subjects : {per} ")
        if (per>=70):
            print("congratulations - You got Distinction")
        elif (per<=70 and per>=40):
            print("congratulations - You got Pass")
        else:
            print("You have Failed")
        print("-------------------------------------------------")
```

#### WAP To Create A class ATM and Create methods 
#### 1> to Genrate a pin
#### 2> to withdraw , deposit money
#### 3> to check balance 
#### 4> to change pin
#### 5> Pogram should only terminate when user wants

```python
class ATM:
    def __init__(self):
        self.amount=10000
        self.genratepin()
        self.choice()
    def genratepin(self):
        self.pin=input("Enter A 4 - digit pin")
        if len(self.pin)>4 and len(self.pin)<4:
            print("Enter a valid 4-digit pin")
            self.genratepin()
        self.choice()
    def withdraw(self,amount):
        pinchk=input("Enter your pin")
        if pinchk==self.pin:
            self.amount-=amount
            print("Money Withdrawed")
            print("Your Balance is %d"%self.amount)
        else:
            print("Enter a valid value")
        self.choice()
    def deposit(self,amount):
        pinchk=input("Enter your pin")
        if pinchk==self.pin:
            self.amount+=amount
            print("Money Deposited")
            print("Your Balance is %d"%self.amount)
        else:
            print("Enter a valid value")
        self.choice()
    def checkbalance(self):
        pinchk=input("Enter your pin")
        if pinchk==self.pin:
            print("Your Balance is %d"%self.amount)
        else:
            print("Enter a valid value")
        self.choice()
    def changepin(self):
        self.genratepin()
        self.choice()
    def choice(self):
        ch=int(input('''
            Enter 1 : TO withdraw money
            Enter 2 : To deposit money
            Enter 3 : TO change pin
            Enter 4 : TO check Balance
            Enter 5 : to Exit\n'''))
        if ch==1:
            self.withdraw(int(input("Enter Money you want to Withdraw")))
        elif ch==2:
            self.deposit(int(input("Enter Money you want to deposit")))
        elif ch==4:
            self.checkbalance()
        elif ch==3:
            self.changepin()
        elif ch==5:
            print("System existing....")
            exit()
            quit()
        else:
            print("Enter a valid value")
        self.choice()
```

#### WAP TO CREATE A CLASS ANALYSIS to analysis any given statement it should have methods to count 
#### 1>number of characters
#### 2> number of uppercase alphabets
#### 3> number of lowercase alphabet
#### 4> number of vowels
#### 5> number of consonats
#### 6> a method to display statistics 
#### it should provide a detailed answer in percenatge

```python
class Analysis:
    def __init__(self):
        self.Stmt=input("Enter Statement : ")
    def upperCS(self):
        self.countUpper=0
        st=(self.Stmt).split()
        for i in st:
            for j in i:
               if j.isupper():
                   self.countUpper+=1
    def lowerCS(self):
        self.countLower=0
        st=(self.Stmt).split()
        for i in st:
            for j in i:
               if j.islower():
                   self.countLower+=1
    def isVowel(self):
        self.countVowel=0
        st=(self.Stmt).split()
        for i in st:
            for j in i.lower():
               if j in ('a','e','i','o','u') and j.isalpha():
                   self.countVowel+=1
    def isConosnats(self):
        self.countCons=0
        st=(self.Stmt).split()
        for i in st:
            for j in i.lower():
               if j.isalpha() and j not in ('a','e','i','o','u'):
                   self.countCons+=1
    def ttlChar(self):
        self.countCh=0
        st=self.Stmt
        for i in st:
            for j in i:
               self.countCh+=1
    def stats(self):
        print("---------------------------------------------------------------")
        print(f"Total number of characters : {self.countCh}\n")
        print(f"Total number of vowels : {self.countVowel}\n")
        print(f"Total number of consonants : {self.countCons}\n")
        print(f"Total number of Upper case characters : {self.countUpper}\n")
        print(f"Total number of lower case characters : {self.countLower}\n")
        print(f"stats --- percenatge of vowels {round((self.countVowel/self.countCh)*100,2)}")
        print(f"      --- percenatge of consonants {round((self.countCons/self.countCh)*100,2)}")
        print(f"      --- percenatge of upper case alphabets {round((self.countUpper/self.countCh)*100,2)}")
        print(f"      --- percenatge of lower case alphabets {round((self.countLower/self.countCh)*100,2)}")
```

```python
A=Analysis()
A.upperCS()
A.lowerCS()
A.isVowel()
A.isConosnats()
A.ttlChar()
A.stats()
```
